```python
import os
import shutil

# Directory containing files
source_dir = "/content/drive/MyDrive/MIT 805/Assign 2 _attempt 2/Data/Summer"
destination_base_dir = "/content/drive/MyDrive/MIT 805/Assign 2 _attempt 2/Data/Summer"

# Number of files per batch
batch_size = 2000
#batch_size = 500

files = os.listdir(source_dir)

# Partition files into batches
for i in range(0, len(files), batch_size):
    # Create new batch folder
    batch_dir = os.path.join(destination_base_dir, f"batch_{i//batch_size + 1}")
    os.makedirs(batch_dir, exist_ok=True)

    # Move batch of files into new folder
    for file in files[i:i + batch_size]:
        shutil.move(os.path.join(source_dir, file), os.path.join(batch_dir, file))

print("Files have been partitioned into batches.")
```

```python
import os

BASE_DIR = "/content/drive/MyDrive/MIT 805/Assign 2 _attempt 2/Data/Summer"

DEST_DIR = "/content/drive/MyDrive/MIT 805/Assign 2 _attempt 2/Data/Summer/combined/"
os.makedirs(DEST_DIR, exist_ok=True)

for batch in os.listdir(BASE_DIR):
    batch_path = os.path.join(BASE_DIR, batch)

    if os.path.isdir(batch_path) and batch.startswith("batch_"):
        # Create a combined file for each batch
        combined_file = os.path.join(DEST_DIR, f"{batch}.txt")

        with open(combined_file, 'w') as outfile:
            # Combine files with a newline after each file
            for file_name in os.listdir(batch_path):
                file_path = os.path.join(batch_path, file_name)
                if os.path.isfile(file_path):
                    with open(file_path, 'r') as infile:
                        outfile.write(infile.read())
                        outfile.write('\n')

        print(f"Combined files in {batch} into {combined_file}")
```

```python
# Undo moves

import os
import shutil

batch_dir_base = "/content/drive/MyDrive/MIT 805/Assign 2 _attempt 2/Data/Summer"

main_dir = "/content/drive/MyDrive/MIT 805/Assign 2 _attempt 2/Data/Summer"

for batch_dir in os.listdir(batch_dir_base):
    batch_dir_path = os.path.join(batch_dir_base, batch_dir)

    if os.path.isdir(batch_dir_path) and batch_dir.startswith("batch_"):

        for file in os.listdir(batch_dir_path):
            shutil.move(os.path.join(batch_dir_path, file), os.path.join(main_dir, file))


print("Files have been moved back to the main folder.")
```
